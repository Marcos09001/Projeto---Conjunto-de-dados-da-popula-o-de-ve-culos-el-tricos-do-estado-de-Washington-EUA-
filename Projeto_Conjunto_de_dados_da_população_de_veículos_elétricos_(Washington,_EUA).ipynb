# Entendimento do Projeto e Importação de Dados

```python
#Importei as bibliotecas a serem utilizadas no projeto
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
```

```python
df.info()

df.describe()
```

# Limpeza e Preparação dos Dados

```python
#Verificação de duplicatas

df.duplicated().sum()
```

```python
#Remoção de duplicatas

df.drop_duplicates(inplace=True)
```

```python
df
```

```python
#Verificar se possui duplicatas

df.duplicated().sum()
```

```python
#Verificar possíveis valores ausentes/nulos no dataset

df.isnull().sum()
```

```python
#Verificação de NaN

df.isna().sum()
```

```python
#Tratamento das colunas com valores nulos/ausentes

#Substituição de valores por colunas
df.fillna({'County': 'Unknown'}, inplace=True)
df.fillna({'City': 'Unknown'}, inplace=True)
df.fillna({'State': 'Unknown'}, inplace=True)
df.fillna({'Postal Code': 0}, inplace=True)
df.fillna({'Model Year': 0}, inplace=True)
df.fillna({'Make': 'Unknow'}, inplace=True)
df.fillna({'Model': 'Unknow'}, inplace=True)
df.fillna({'Electric Vehicle Type': 'Unknown'}, inplace=True)
df.fillna({'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'Unknown'}, inplace=True)
df.fillna({'Electric Range': 0}, inplace=True)
df.fillna({'Legislative District': 'Unknown'}, inplace=True)
df.fillna({'DOL Vehicle ID': 'Unknown'}, inplace=True)
df.fillna({'Vehicle Location': 'Unknown'},inplace=True)
df.fillna({'Electric Utility': 'Unknown'}, inplace=True)
df.fillna({'2020 Census Tract': 0}, inplace=True)
```

```python
#Verificar se ainda restam valores nulos/ausentes nas colunas

df.isna().sum()
```

```python
df
```

```python
# Base de dados limpa

df.to_csv('df_clean.csv')
```

# Análise Exploratória dos Dados

```python
# Contagem de carros eletricos por Cidade

count_city = df['City'].value_counts()
count_city
```

```python
# Contagem de carros eletricos por Estado

count_state = df['State'].value_counts()
count_state
```

```python
# Contagem de carros por Modelo

model_count = df['Model'].value_counts()
model_count
```

```python
# Contagem de carros por fabricante

make_count = df['Make'].value_counts()
make_count

# Exportar para CSSV

make_count.to_csv('make_count.csv')

make_count
```

```python
# Contagem de tipo de utilitario eletrico

type_eletric = df['Electric Utility'].value_counts()
type_eletric
```

```python
# Relação entre modelo de carro e ano de modelo

model_year = df.groupby('Model')['Model Year'].value_counts()
model_year
```

```python
# Relação entre fabricante e modelos

make_model = df.groupby('Make')['Model'].value_counts()
make_model

# Covertendo para csv

make_model.to_csv('make_model.csv')

make_model
```

```python
# Autonomia maxima de cada modelo

max_range = df.groupby('Model')['Electric Range'].max()
max_range

# Convertendo para csv

max_range.to_csv('max_range.csv')

max_range
```

```python
# Media de autonomia por modelo de carro

model_mean = df.groupby('Model')['Electric Range'].mean()
model_mean.round(2)
```
